```python
# IMPORTS
import json
import nltk
import numpy as np
import pandas as pd
import textwrap
import random


# DATA
with open("texts.json", 'r') as f:
    texts = json.load(f)

# Check to see that the data loaded as expected
texts[30:35]
# This uses standard libraries (no imports needed)
from io import StringIO
from html.parser import HTMLParser

# FUNCTIONS
# From https://stackoverflow.com/questions/753052/strip-html-from-strings-in-python (edited)

class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.text = StringIO()
    def handle_data(self, d):
        self.text.write(d)
    def get_data(self):
        return self.text.getvalue()

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()
```

```python
# CLEANUP

# Remove the HTML
plaintexts = [strip_tags(i) for i in texts]
# Change ampersands to "and"s
plaintexts = [x.replace("&amp;", "and") for x in plaintexts]
plaintexts = [y.replace("\n", ".") for y in plaintexts]

# Tokenize all the texts
gabs = [ nltk.tokenize.word_tokenize(i.lower()) for i in plaintexts ]

# Make one composite list
workable = [ i for i in gabs if 20 < len(i) < 500 ]
```

```python
# Let's sample 3 texts from the longer list:
for i in random.sample(workable, 3):
    s = " ".join(i)
    print('\n'.join(textwrap.wrap(s, width=80, replace_whitespace=False)))
    print("\n")
```
